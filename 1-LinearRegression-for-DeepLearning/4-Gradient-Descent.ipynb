```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
N = 10
D = 3
X = np.zeros((N,D))

X[:,0] = 1 #bias term
X[:5,1] = 1 #第2列前5行全部为1
X[5:,2] = 1 #第3列后5行全部为1

Y = np.array([0]*5 + [1]*5)
```

```python
# gradient descent

costs = [] # keep track of squared error cost

w = np.random.randn(D)/np.sqrt(D) # randomly initialize w
learning_rate = 0.001

for t in range(1000):
    Yhat = X.dot(w)
    w = w -learning_rate*X.T.dot(Yhat - Y)
    cost = (Y - X.dot(w)).T.dot(Y - X.dot(w))
    costs.append(cost)
```

```python
plt.plot(costs)
plt.show()
```

```python
print("final w:", w)
```

```python
# plot prediction and actual target

plt.plot(Yhat,label='prediction')
plt.plot(Y,label='target')
plt.legend()
plt.show()
```
