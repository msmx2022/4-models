```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
path = "C:\\Users\\A25563592\\Desktop\\DataAnalysis\\03_Analysis\\msmx-github\\4-models\\1-LinearRegression-for-DeepLearning\\data_1d.csv"
ColNames = ['X','Y']
df = pd.read_csv(filepath_or_buffer=path,header=None,names=ColNames)
```

```python
df.dtypes
```

```python
plt.scatter(df['X'],df['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# y_hat = ax + b

X = df['X']
Y = df['Y']

a = (X.dot(Y) - Y.mean()*X.sum())/(X.dot(X)-X.mean()*X.sum())
b = (Y.mean()*X.dot(X) - X.mean()*X.dot(Y))/(X.dot(X)-X.mean()*X.sum())

Y_hat = a*X + b
```

```python
plt.scatter(df['X'],df['Y'])
plt.plot(X,Y_hat)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# R_squared

#R_squared = 1- Residual_y_hat / Residual_y_total

Residual_y_hat = (Y - Y_hat).dot((Y - Y_hat))

Residual_y_total = (Y - Y.mean()).dot((Y - Y.mean()))

R_squared = 1- Residual_y_hat / Residual_y_total


# R_squared  1Residual_y_hat almost be zero, forecast  actual

# R_squared  0y_hat is not better than y_mean()

```

```python
path = "C:\\Users\\A25563592\\Desktop\\DataAnalysis\\03_Analysis\\msmx-github\\4-models\\1-LinearRegression-for-DeepLearning\\data_2d.csv"
ColNames = ['X1','X2','Y']
df = pd.read_csv(filepath_or_buffer=path,header=None,names=ColNames)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(df['X1'],df['X2'],df['Y'])
plt.show()
```

```python
X = np.array(df[['X1','X2']])

one_column = np.ones((X.shape[0],1))

X_mdf = np.hstack((one_column,X))

Y = np.array(df['Y'])

```

```python
# 
W = np.linalg.solve((X_mdf.T).dot(X_mdf),(X_mdf.T).dot(Y))
W
```

```python
#Y_hat
Y_hat = X_mdf.dot(W)
```

```python
# R_squared

Residual_yhat = (Y - Y_hat).dot(Y - Y_hat)

Residual_ymean = (Y - Y.mean()).dot(Y - Y.mean())

R_squared = 1- Residual_yhat/Residual_ymean

R_squared
```
